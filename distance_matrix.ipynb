```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
```

```python
def get_distance_time(origin, destination):
    url = "https://maps.googleapis.com/maps/api/distancematrix/json"
    params = {
        "origins": origin,
        "destinations": destination,
        "key": 'ha ha lol'
    }

    response = requests.get(url, params=params)
    data = response.json()

    if response.status_code == 200:
        distance = data['rows'][0]['elements'][0]['distance']['text']
        duration = data['rows'][0]['elements'][0]['duration']['text']
        
        return distance, duration
    else:
        print("Error:", response.status_code)
        return None, None
```

```python
def duration_to_decimal(duration_str):
    
    parts = duration_str.split()
    hours = 0
    minutes = 0
    for i in range(len(parts)):
        if parts[i] == 'hour' or parts[i] == 'hours':
            hours = int(parts[i-1])
        elif parts[i] == 'min' or parts[i] == 'mins':
            minutes = int(parts[i-1])
    
    decimal_duration = hours + minutes / 60
    return decimal_duration
```

```python

duration_matrix.fillna(0, inplace=True)
duration_matrix =duration_matrix.astype(str)
duration_matrix.index = duration_matrix.index.str.replace(", UK", "")
duration_matrix.columns = duration_matrix.columns.str.replace(", UK", "")

for col in duration_matrix.columns:
    duration_matrix[col] = duration_matrix[col].apply(duration_to_decimal).round(2)

duration_matrix
```
