```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Distance_1 = pd.read_csv('Distance_cleaned.csv', index_col=0)
Duration_1 = pd.read_csv('Duration_cleaned.csv', index_col=0)  
trips = pd.read_csv('no_of_trips.csv', index_col=0)
trips.drop("Total", axis = 1, inplace = True) 
trips
```

```python
Distance = Distance_1
plt.figure(figsize=(10, 8))
sns.heatmap(Distance, cmap='viridis', annot=False, fmt=".1f", linewidths=.5, )
plt.title('Distance between Towns')
plt.xlabel('Destination')
plt.ylabel('Origin')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(trips, cmap='viridis', annot=False, fmt=".1f", linewidths=.5, vmin=470, vmax=20000)
plt.title('Goods moved between Locations (Model Weightings)')
plt.xlabel('Destination')
plt.ylabel('Origin')
plt.show()
```
