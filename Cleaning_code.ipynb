```python


Distance.fillna(0, inplace=True)
Distance =Distance.astype(str)
for col in Distance.columns:
    Distance[col] = Distance[col].str.replace(" km", "")
Distance.index = Distance.index.str.replace(", UK", "")
Distance.columns = Distance.columns.str.replace(", UK", "")
Distance.columns
```

```python
def duration_to_decimal(duration_str):
    
    parts = duration_str.split()
    hours = 0
    minutes = 0
    for i in range(len(parts)):
        if parts[i] == 'hour' or parts[i] == 'hours':
            hours = int(parts[i-1])
        elif parts[i] == 'min' or parts[i] == 'mins':
            minutes = int(parts[i-1])
    
    decimal_duration = hours + minutes / 60
    return decimal_duration

```

```python

Duration.fillna(0, inplace=True)
Duration =Duration.astype(str)
Duration.index = Duration.index.str.replace(", UK", "")
Duration.columns = Duration.columns.str.replace(", UK", "")

for col in Duration.columns:
    Duration[col] = Duration[col].apply(duration_to_decimal).round(2)

Duration
```
